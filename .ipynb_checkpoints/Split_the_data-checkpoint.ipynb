```python
import pandas as pd
from sklearn.model_selection import train_test_split

def split_the_data(file_path, target_columns, name_plus, path_to) :
        # Load the data from the 'data_processed.csv' file 
    df = pd.read_csv(file_path, low_memory=False)

    # Split the data into X and Y
    
    all_columns = set(df.columns)
    target_columns_set = set(target_columns)
    X_columns = list(all_columns - target_columns_set)
    
    X = df[X_columns]  # Features from the beginning up to "Story Area"
    Y = df[target_columns]  # 

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    file_path_X_train = path_to +  'X_train_' + name_plus + '.csv'
    X_train.to_csv(file_path_X_train, index=False)

    file_path_X_test = path_to + 'X_test_' + name_plus + '.csv'
    X_test.to_csv(file_path_X_test, index=False)

    file_path_Y_train = path_to + 'Y_train_' + name_plus + '.csv'
    Y_train.to_csv(file_path_Y_train, index=False)

    file_path_Y_test = path_to + 'Y_test_' + name_plus + '.csv'
    Y_test.to_csv(file_path_Y_test, index=False)

#Split the data for C part 1 before FE
path_from = './Files/Before_Feature_Engineering'
path_to = 'Files/Before_Feature_Engineering/Split/'
file_path_C_part1 = path_from + '/data_C_part1.csv'
target_column_C = ["Nail spacing [cm]", "Number sheathing panels", "Number end studs", "Total number studs","HoldDown Model / ATS"]
df_C_part1 = split_the_data(file_path_C_part1, target_column_C, 'C_part1', path_to) 

#Split the data for C part 2 before FE
file_path_C_part2 = path_from + '/data_C_part2.csv'
target_column_C_2 = ['Tx(s)', 'Ty(s)']
df_C_part2 = split_the_data(file_path_C_part2, target_column_C_2, 'C_part2', path_to) 

#Split the data for D before FE
file_path_D = path_from + '/data_D.csv'
target_column_D = ['Ωx', 'Ωy', 'µx', 'µy', 'CMR', 'SSF', 'ACMR', 'IO-ln θ','IO-β','LS-ln θ','LS-β', 'CP-ln θ','CP-β']
df_D = split_the_data(file_path_D, target_column_D, 'D', path_to) 
    

path_from_FE = './Files/After_Feature_Engineering'
path_to_FE = 'Files/After_Feature_Engineering/Split/'
#Split the data for C part 1 after FE
file_path_C_part1 = path_from_FE + '/data_C_part1_FE.csv'
df_C_part1 = split_the_data(file_path_C_part1, target_column_C, 'C_part1_FE', path_to_FE) 

#Split the data for C part 2 after FE
file_path_C_part2 = path_from_FE + '/data_C_part2_FE.csv'
df_C_part2 = split_the_data(file_path_C_part2, target_column_C_2, 'C_part2_FE', path_to_FE) 

#Split the data for D after FE
file_path_D = path_from_FE + '/data_D_FE.csv'
df_D = split_the_data(file_path_D, target_column_D, 'D_FE', path_to_FE) 
```
